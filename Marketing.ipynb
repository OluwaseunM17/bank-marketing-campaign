# Data Cleaning Commences from here onwards.

```python
# Splitting 1 string column into multiple columns
data[["age", "job", "marital", "education", "default", "balance",
"housing","loan", "contact", "day", "month", "duration","campaign",
"pdays", "previous", "poutcome","y"]] = data['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(";",expand=True)
```

```python
# Dropping Unneccesary Column seeing as the required info has been obtained
data.drop('age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"', axis=1, inplace=True)
```

```python
data.info()
```

```python
# to remove the quotation marks on the strings
data["job"] = data["job"].str.replace('.','')
data["job"] = data["job"].str.replace('"','')
data["job"] = data["job"].str.strip()
data["marital"] = data["marital"].str.extract(r'(\w+)')
data["education"] = data["education"].str.extract(r'(\w+)')
data["default"] = data["default"].str.extract(r'(\w+)')
data["housing"] = data["housing"].str.extract(r'(\w+)')
data["loan"] = data["loan"].str.extract(r'(\w+)')
data["contact"] = data["contact"].str.extract(r'(\w+)')
data["month"] = data["month"].str.extract(r'(\w+)')
data["poutcome"] = data["poutcome"].str.extract(r'(\w+)')
data["y"] = data["y"].str.extract(r'(\w+)')
```

```python
def num(x):
    data[x] = data[x].astype(int)
```

```python
#to change the data type of digits to integers
numb = ["age", "balance", "day", "duration", "campaign", "pdays", "previous"]
for x in numb:
    num(x)
```

```python
# to change the data type of the strings to categories; this is done for 18,19,20.
loe = ["unknown","secondary","primary","tertiary"]
data["education"] = pd.Categorical(data["education"], categories=loe)
```

```python
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
data['month'] = pd.Categorical(data['month'], categories=month_order, ordered=True)
```

### A Break down of the subscription rates by month using a stacked area plot

```python
# Creating columns in original dataframe  and creating another dataframe to enable some calculations to come.
data['is_yes'] = data['y'] == 'yes'
monthly_subscriptions = data.groupby('month').agg({'is_yes': 'sum', 'y': 'count'}).reindex(month_order)
monthly_subscriptions['subscription_rate'] = monthly_subscriptions['is_yes'] / monthly_subscriptions['y']

# Prepare data for stacked area plot
stacked_data = pd.DataFrame({
    'Month': month_order,
    'Yes': monthly_subscriptions['is_yes'],
    'No': monthly_subscriptions['y'] - monthly_subscriptions['is_yes']
}).set_index('Month')

# Create the stacked area plot
plt.figure(figsize=(10, 6))
stacked_data.plot(kind='area', stacked=True, alpha=0.6)
plt.title('Subscription Rates by Month')
plt.xlabel('Month')
plt.ylabel('Number of Subscriptions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Insights:
Analysis from the dataset revealed the following:
- The dataset revealed that there are 4521 customers and only 521 customers subsscribed for the term loan that the bank was advertising and that brings the campaigns subscription rate to about 11.52%
- The Average contact duration of duration of the bank to every customer is approximately 264 seconds (around 4 and a half minutes).
- The total number of times the bank reached out to their customers is 12630 times throughout the year (From January to December)
- The months with the highest number of times for customer contacts is in May whereas the month with the lowest number of times for customer contacts is in in December.
- The months with the highest number of subscribed customers is in May whereas the month with the lowest number of subscribed numbers is in December.
- The Job role with the highest subscription rate was from the category of people who are retired and then those with the lowest are those from the blue collar category.
- A bulk of the customers that the marketing campaign reached out to are within the ages of around 26-59.
There were 6 age groups noticed according to the periods they were born in and these age groups are as follows: genz, millenials, genx, genj, boomers1, postwar.
- From the table above Postwar age (79-96) had the highest subscription rate whereas the millenials (28-43) had the lowest subscription rate but one should also bear in mind that these results are relative seeing as the number of customers in the different age groups differ.
- Based on Level of education we had those who had finished from tertiary institutions with the highest subscription rate and those  who from primary institutions with the lowest subscription rates but we also must bear in mind that this is also relative.
- In terms of contribution of age groups to the overall subscription rate the group that contributed the most to the overall subscription rate was the millenials age group and the lowest contribution came from the postwar age group.
- In terms of contribution of levels of education to the overall subscription rate the group that contributed the most to the overall subscription rate was the secondary and the lowest was from those who we couldnt account account for their level of education(Unknown).    

# Recommendations 
- A correlation was observed between the number of times a customer was contacted and the subscription success, so it might be a good idea to explore being more intentional towards reaching out to a customer but bear in mind that this shouldnt be the main strategy seeing as this might increase the rate of subscription but not as wanted
- Reaching out to Aged customers(60+) should be a strategy that should be explored cause the subscription rate for the older generation seemed better than that of much younger customers.
- Reaching out & campaigning to those who finished from tertiary institutions should be another strategy thats investigated to improve the campaign performance.
- Although this is similar to the 2nd point but then reaching out to those who are retired should also be considered cause sometimes people can be old but not retired yet.
- The lowest performing Job roles were those in blue collar jobs and then considering this, maybe in campaigning the firm can market ideas that would be more suitable to people in the blue collar category i.e. marketing more attractive ideas to the people of this group and this would require more research.
